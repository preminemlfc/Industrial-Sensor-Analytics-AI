The Project: Industrial Machine Sensor Analysis

```python
import sqlite3
import pandas as pd
import numpy as np
```

Creating a virtual database with 1000 rows of sensor data

```python
# 1. Connect to SQLite (creates the file automatically)
conn = sqlite3.connect('factory_data.db')
cursor = conn.cursor()

# 2. Create a Table for Sensor Readings
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sensors (
        id INTEGER PRIMARY KEY,
        machine_id TEXT,
        voltage REAL,
        temperature REAL,
        status TEXT
    )
''')

# 3. Generate Dummy Data (1000 rows)
machines = ['Machine_A', 'Machine_B', 'Machine_C']
data = []
for i in range(1000):
    data.append((
        np.random.choice(machines),
        np.random.uniform(210, 240), # Voltage
        np.random.uniform(40, 95),   # Temp in Celsius
        np.random.choice(['Active', 'Fault', 'Idle'], p=[0.8, 0.1, 0.1])
    ))

cursor.executemany('INSERT INTO sensors (machine_id, voltage, temperature, status) VALUES (?, ?, ?, ?)', data)
conn.commit()
print("Database 'factory_data.db' created with 1,000 sensor readings!")
```

#To Find the average temperature for each machine, but only for those currently in 'Fault' status."

```python
query = """
    SELECT machine_id, 
           AVG(temperature) as avg_temp, 
           COUNT(*) as fault_count
    FROM sensors
    WHERE status = 'Fault'
    GROUP BY machine_id
    HAVING avg_temp > 60
"""

df_analysis = pd.read_sql_query(query, conn)
print("\n--- Analytics Insight: Machines with High-Temp Faults ---")
print(df_analysis)
```

To turn our Findings into Visual Insight

```python
import matplotlib.pyplot as plt

# 5. Simple Plot: Temp vs Voltage
df_all = pd.read_sql_query("SELECT voltage, temperature FROM sensors WHERE status='Active'", conn)

plt.scatter(df_all['voltage'], df_all['temperature'], alpha=0.5, color='blue')
plt.title('Industrial Sensor Health: Temperature vs Voltage')
plt.xlabel('Voltage (V)')
plt.ylabel('Temperature (C)')
plt.show()

conn.close()
```
